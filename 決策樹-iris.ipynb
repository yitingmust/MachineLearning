```python
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 讀入鳶尾花資料
iris = load_iris()
iris_data = iris.data  # 特徵資料
iris_label = iris.target  # 目標值

# 切分訓練與測試資料
train_data, test_data, train_label, test_label = train_test_split(iris_data, iris_label, test_size = 0.3)

# 建立分類器
clf = tree.DecisionTreeClassifier()
iris_clf = clf.fit(train_data, train_label) # 使用訓練資料進行模型訓練

print("訓練資料正確率:",clf.score(train_data,train_label)) # 訓練資料正確率
print("測試資料正確率:",clf.score(test_data, test_label)) # 測試資料正確率

predict=clf.predict(test_data) # 預測
print("標準答案:",test_label) # 標準答案
print("預測答案:",predict) # 預測答案
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None) # 將決策樹模型轉換為graphviz格式的數據
graph = graphviz.Source(dot_data) # 將graphviz格式的數據轉換為可視化圖形
graph.render("iris") #將結果存成pdf檔案
```

```python
import pydotplus
from IPython.display import Image
dot_data = tree.export_graphviz(clf, out_file=None,
                feature_names=iris.feature_names, # 特徵名稱
                class_names=iris.target_names,  # 目標類別名稱
                filled=True, rounded=True,  # 圖形填充和邊角處理
                special_characters=True)  # 特殊字元處理
graph = pydotplus.graph_from_dot_data(dot_data)  # 將graphviz格式的數據轉換為可視化圖形
Image(graph.create_png())  #將結果用圖形化(PNG)呈現
```
